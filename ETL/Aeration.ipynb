```python
# Import dependencies
import pandas as pd
from sqlalchemy import create_engine
import time
```

```python
# Connecting to the Database
engine = create_engine("postgresql://postgres:postgres@localhost/WWTP")
conn = engine.connect()

# Reading SQL query into a Dataframe 
df_temp1 = pd.read_sql_query('select * from "Aeration_temp_1"', con=conn)
df_temp2 = pd.read_sql_query('select * from "Aeration_temp_2"', con=conn)
```

```python
# Merging Dataframes
df = pd.merge(df_temp1, df_temp2, on='Time', how='outer')
df
```

```python
# Write to PostgreSQL
start_time = time.time()
df.to_sql('Aeration', con=conn, if_exists='replace', index=False)
print("to_sql duration: {} seconds".format(time.time() - start_time))
```

```python
# Close the connection
conn.close()
```
