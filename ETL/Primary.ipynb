```python
# Import dependencies
import pandas as pd
from sqlalchemy import create_engine
import time
```

```python
# Connecting to the Database
engine = create_engine("postgresql://postgres:postgres@localhost/WWTP")
conn = engine.connect()

# Reading SQL query into a Dataframe 
df_temp1 = pd.read_sql_query('select * from "temp_1"', con=conn)
df_temp2 = pd.read_sql_query('select * from "temp_2"', con=conn)
df_temp3 = pd.read_sql_query('select * from "primary_temp_1"', con=conn)
df_temp4 = pd.read_sql_query('select * from "primary_temp_2"', con=conn)
df_temp5 = pd.read_sql_query('select * from "primary_temp_3"', con=conn)
df_temp6 = pd.read_sql_query('select * from "primary_temp_4"', con=conn)
df_temp7 = pd.read_sql_query('select * from "primary_temp_5"', con=conn)
df_temp8 = pd.read_sql_query('select * from "primary_temp_6"', con=conn)
df_temp9 = pd.read_sql_query('select * from "primary_temp_7"', con=conn)
df_temp10 = pd.read_sql_query('select * from "primary_temp_8"', con=conn)
```

```python
# Selecting specific columns
df_temp1 = df_temp1[['Time','T1_2_3_4-S1_2-PRI-INF-Flow','T5_6-S3-PRI-INF-Flow']]
```

```python
# Selecting specific columns
df_temp2 = df_temp2[['Time', 'T1-S1-PRI-P1-RSD', 'T1-S1-PRI-P2-RSD', 'T2-S1-PRI-P3-RSD',
       'T2-S1-PRI-P4-RSD', 'T3-S2-PRI-P5-RSD', 'T3-S2-PRI-P6-RSD',
       'T4-S2-PRI-P7-RSD', 'T4-S2-PRI-P8-RSD', 'T5-S3-PRI-P9-RSD',
       'T5-S3-PRI-P10-RSD', 'T5-S3-PRI-P11-RSD', 'T6-S3-PRI-P12-RSD',
       'T6-S3-PRI-P13-RSD', 'T6-S3-PRI-P14-RSD']]
```

```python
# Merging Dataframes
df_1 = pd.merge(df_temp1, df_temp2, on='Time', how='outer')
df_2 = pd.merge(df_1, df_temp3, on='Time', how='outer')
df_3 = pd.merge(df_2, df_temp4, on='Time', how='outer')
df_4 = pd.merge(df_3, df_temp5, on='Time', how='outer')
df_5 = pd.merge(df_4, df_temp6, on='Time', how='outer')
df_6 = pd.merge(df_5, df_temp7, on='Time', how='outer')
df_7 = pd.merge(df_6, df_temp8, on='Time', how='outer')
df_8 = pd.merge(df_7, df_temp9, on='Time', how='outer')
```

```python
# Merging Dataframes
df = pd.merge(df_8, df_temp10, on='Time', how='outer')
df
```

```python
# Write to PostgreSQL
start_time = time.time()
df.to_sql('Primary', con=conn, if_exists='replace', index=False)
print("to_sql duration: {} seconds".format(time.time() - start_time))
```

```python
# Close the connection
conn.close()
```
